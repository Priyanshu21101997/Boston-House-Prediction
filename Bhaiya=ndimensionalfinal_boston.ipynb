```python
import numpy as np
data = np.loadtxt("train.csv",delimiter=",")
xx = data[:,0:13]
a = np.ones(len(xx),int).reshape(-1,1)
x = np.hstack((xx,a))


y = data[:,13]

test=np.loadtxt("Test.csv",delimiter=",")
x_test=test[:,0:13]

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
xx=sc.fit_transform(xx)
X = sc.transform(x_test)



def cost(x, y, m):
    N = len(x)
    cost = 0
    for i in range(N):
        y_p = 0
        for j in range(x.shape[1]):
            y_p += x[i][j] * m[j] 
        cost += (1/N) * (y[i] - y_p) ** 2
    return cost
    
def ans(x,y):
#     m=c=0
    m = np.zeros(x.shape[1])
    learning_rate=0.1
    iterations=3790
    for k in range (iterations):
        m_slope = np.zeros(x.shape[1])
        N = len(x)
        for i in range(len(m_slope)):
            for j in range(N):
                y_p =(m * x[j]).sum()
                m_slope[i] += (-2/N) * (y[j] - y_p ) * x[j][i] 

        m= m - learning_rate * m_slope
       
#         dm=((-2/n)*(y-m*x-c)*x).sum()
#         dc=((-2/n)*(y-m*x-c)).sum()
#         m=m-learning_rate*dm
#         c=c-learning_rate*dc
        co=cost(x,y,m)

        print("M:",m,'\t',"Cost:",co,'\t',"Iteration:",k)
    #return m

M=ans(xx,y)

Y=[]
for i in range (len(X)):
    Y.append((M[0:13]*X[i]).sum()+M[13])
save=np.savetxt("final.csv",Y,delimiter=",",fmt='%1.10f')
```
